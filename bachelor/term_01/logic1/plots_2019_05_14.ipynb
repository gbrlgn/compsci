```
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0.0,4.0,0.01)
y = 1/(1 + np.exp(-x))
x2 = np.arange(-0.0,-4.0,-0.01)
y2 = 1/(1 + np.exp(x))

print(len(x))
print(len(y))

fig, ax = plt.subplots()

ax.plot(x,y,"-",x2,y2,"-")

print("Fim")
```

```
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0.0,4.0,0.01)
y = (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))
x2 = np.arange(-0.0,-4.0,-0.01)
y2 = (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))

print(len(x))
print(len(y))

fig, ax = plt.subplots()

ax.plot(x,y,"-",x2,y2,"-")

print("Fim")
```

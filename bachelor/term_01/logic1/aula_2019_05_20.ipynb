```
#Gráfico 01

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-10., 10., 0.001) #gerando o valor de x
y = -((x**2) - (2.5*x)) #implementando a função

#plotar o gráfico

f, c = plt.subplots()
c.plot(x,y)


print(len(y))
```

```
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-10., 10., 0.001)
y = []

for v in x:
  result = 0 #guarda o valor da equação
  
  if v <= -2:
    result = (-2*v)-1
  
  if v > -2 and v < 1:
    result = 3
  
  if v >= 1:
    result = (2*v)+1
    
  y.append(result)

f, c = plt.subplots()
c.plot(x,y)


print(len(y))
```

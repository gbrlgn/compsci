```
lst_temp = []
lst_dist = []
lst_velo = []
lst_percent = []
lst_testes = []
lst_mach = []
lst_machDesc = []


def cal_velo():
  flo_velo = int_dist / int_temp
  lst_velo.append(flo_velo)
  
  #porcentagem
  int_percent = (flo_velo * 100) // 343
  lst_percent.append(int_percent)
  
  #velocidade Mach
  flo_mach = flo_velo / 343
  
  if flo_mach < 0.8:
    lst_mach.append(flo_mach)
    lst_machDesc.append("subsônico")
  elif flo_mach > 0.8 and flo_mach < 1.3:
    lst_mach.append(flo_mach)
    lst_machDesc.append("transsônico")
  elif flo_mach > 1.3 and flo_mach < 5.0:
    lst_mach.append(flo_mach)
    lst_machDesc.append("supersônico")
  elif flo_mach > 5.0 and flo_mach < 10.0:
    lst_mach.append(flo_mach)
    lst_machDesc.append("hipersônico")
  elif flo_mach > 10.0 and flo_mach < 25.0:
    lst_mach.append(flo_mach)
    lst_machDesc.append("alto-hipersônico")
  elif flo_mach > 25.0:
    lst_mach.append(flo_mach)
    lst_machDesc.append("velocidade de entrada atmosférica")
    
  lst_testes.append(len(lst_testes) + 1)
    
  
def results():
  for p, n in enumerate(lst_testes):
    print("{}) Tempo decorrido: {} s ~ Distância percorrida: {} m ~ Velocidade: {} m/s ~ {}% da velocidade do som ~ Mach {} ({})".format(n, lst_temp[p], lst_dist[p], lst_velo[p], lst_percent[p], lst_mach[p], lst_machDesc[p]))
   

while True:
  print("Insira os valores necessários.")
  print("Se ambos forem iguais a 0, a execução será finalizada e os resultados serão retornados.")
  int_temp = int(input("Tempo decorrido (em segundos): "))
  int_dist = int(input("Distância percorrida (em metros): "))
  
  
  if int_temp == 0:
    break
  
  elif int_dist == 0:
    break
  
  else:
    lst_temp.append(int_temp)
    lst_dist.append(int_dist)
    cal_velo()
    
results()
 
print("Fim da execução.")
```

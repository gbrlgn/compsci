```
#Gr√°fico 01
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-10.0,10.0,0.001)
y = x**2

#Criando os objetos para plotagem

papel, caneta = plt.subplots()
caneta.plot(x,y)

plt.show()
```

```
#Gr√°fico 02
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-10.0,10.0,0.001)
y = np.sin(x + (2 * np.pi * x))

papel, caneta = plt.subplots()
caneta.plot(x,y)

plt.show()
```

```
#Gr√°fico 03
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-10.0,10.0,0.001)
y = np.cos((np.pi / 6)+(4 * np.pi * x))

papel, caneta = plt.subplots()
caneta.plot(x,y)

plt.show()
```

```
#Gr√°fico 04
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-10.0,10.0,0.001)
y = []
for v in x:
  if v <= 1:
    y.append(1 - v)
  elif v > 1:
    y.append(v**2)
    

papel, caneta = plt.subplots()
caneta.plot(x,y)

plt.show()
```

```
#Gr√°fico 05
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-10.0,10.0,0.001)
y = -(x - 1) - (x + 2)

papel, caneta = plt.subplots()
caneta.plot(x,y)

plt.show()
```
